# Escaladores de Datos en `sklearn`

En la biblioteca `scikit-learn`, se proporcionan varias herramientas para escalar y normalizar datos, lo que es una prctica comn antes de aplicar algoritmos de aprendizaje automtico. A continuacin, se describen cuatro escaladores principales: `MinMaxScaler`, `StandardScaler`, `RobustScaler` y `MaxAbsScaler`.

## MinMaxScaler

`MinMaxScaler` transforma las caractersticas para que estn dentro de un rango especfico, tpicamente entre 0 y 1. Esta tcnica es til cuando queremos garantizar que todas las caractersticas tengan la misma escala.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
````
Ventajas: Mantiene las relaciones entre los valores de los datos.

Desventajas: Sensible a los valores atpicos (outliers).

`````python
from sklearn.model_selection import train_test_split

X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values

X_train, X_test, y_train, y_test =\
    train_test_split(X, y,
                    test_size=0.3,
                    random_state=0,
                    stratify=y)
`````

`````python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
x_train_norm = mms.fit_transform(X_train)
x_test_nrom = mms.transform(X_test)
`````

`````python
print(x_train_norm)
`````

`````python
print(x_test_nrom)
`````

## StandardScaler
StandardScaler estandariza las caractersticas eliminando la media y escalando a la varianza unitaria (es decir, hace que la media de los datos sea 0 y la desviacin estndar sea 1).

Ventajas: Funciona bien para datos que siguen una distribucin aproximadamente normal.

Desventajas: Sensible a los valores atpicos.

`````python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
x_train_std = stdsc.fit_transform(X_train)
x_test_std = stdsc.transform(X_test)
`````

`````python
print(x_train_std)
`````

# RobustScaler
RobustScaler utiliza la mediana y los cuartiles para escalar las caractersticas, lo que lo hace ms robusto frente a los valores atpicos.

Ventajas: Robusto frente a los valores atpicos.

Desventajas: No garantiza un rango especfico para los valores transformados.

`````python
from sklearn.preprocessing import RobustScaler
rbs = RobustScaler()
x_train_robust = rbs.fit_transform(X_train)
x_test_robust = rbs.transform(X_test)
`````

`````python
print(x_test_robust)
`````

## MaxAbsScaler
MaxAbsScaler escala los datos de forma que el valor absoluto mximo de cada caracterstica sea 1. Este escalador es til para datos que ya estn centrados en cero y donde la conservacin de la dispersin original es importante.

Ventajas: No desplaza/centra los datos, lo que puede ser til para datos dispersos con muchos ceros.

Desventajas: Menos efectivo si los datos no estn ya aproximadamente centrados en cero.

`````python
from sklearn.preprocessing import MaxAbsScaler
mas = MaxAbsScaler()
x_train_maxabs = mas.fit_transform(X_train)
x_test_maxabs = mas.transform(X_test)
`````

## Comparacin de Escaladores

| Escalador        | Sensible a Outliers | Centro en Cero | Rango Comn | Uso Comn                    |
|------------------|---------------------|----------------|-------------|-----------------------------|
| MinMaxScaler     | S                  | No             | [0, 1]      | Normalizacin simple        |
| StandardScaler   | S                  | S             | No          | Datos con distribucin normal|
| RobustScaler     | No                  | S             | No          | Datos con outliers          |
| MaxAbsScaler     | No                  | No             | [-1, 1]     | Datos dispersos con ceros   |

`````python
def boxplot_iris(feature_names, dataset):
    i=1
    plt.figure(figsize=(11,9))
    for col in feature_names:
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom=False)
        dataset[col].plot(kind='box', subplots=True, sharex=False)
        plt.title(col)
        i += 1
    plt.show()
`````

`````python
boxplot_iris(iris.columns[:-1], iris)
`````

`````python
def histogram_iris(feature_names, dataset):
    i=1
    plt.figure(figsize=(11,2))
    for col in feature_names:
        plt.subplot(1,4,i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom=False)
        dataset[col].plot(kind='hist', subplots=True, color="blue", grid=True)
        plt.title(col)
        i += 1
    plt.show()
`````

`````python
histogram_iris(iris.columns[:-1], iris)
`````

`````python
corr=iris.corr(numeric_only=True)
cmap=sns.diverging_palette(220,20, as_cmap=True)
plt.figure(figsize=(11,9))
sns.heatmap(corr, cmap=cmap, vmax=1.0, vmin=-1.0, center=0, square=True,
            linewidths=.5, cbar_kws={"shrink":.5})
plt.show()
`````

`````python
def piechar_iris(feature_names, target, dataset):
    i=1
    plt.figure(figsize=(11,9))
    for colName in [target]:
        labels=[];sizes=[];
        df=dataset.groupby(colName).size()
        for key in df.keys():
            labels.append(key)
            sizes.append(df[key])
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both', left=True,
                        top=False, right=False,
                        bottom=True, labelleft=True,
                        labeltop=True, labelright=False,
                        labelbottom=False)
        plt.pie(sizes, labels=labels, autopct='%1.1f%%',
                shadow=True, startangle=140)
        plt.axis('equal')
        i += 1
    plt.show()
`````

`````python
piechar_iris(iris.columns[:-1], iris.species, iris)
`````

`````python
from sklearn.model_selection import train_test_split
`````

`````python
X_train, X_test, y_train, y_test=train_test_split(iris.iloc[:,:-1], iris.iloc[:,-1],
                                                  test_size=0.33, random_state=42)
`````

`````python
from sklearn.tree import DecisionTreeClassifier
`````

`````python
model = DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=None,min_samples_split=2,
                               min_samples_leaf=1,min_weight_fraction_leaf=0.0,max_features=None,
                               random_state=42,max_leaf_nodes=None,min_impurity_decrease=0.0,
                               class_weight=None)
`````

`````python
model.fit(X_train, y_train)
`````

`````python
model.score(X_test, y_test)
`````

`````python
import numpy as np
from sklearn.model_selection import cross_val_score, KFold
cv=KFold(n_splits=10,shuffle=True,random_state=42)
results=cross_val_score(model, X_train, y_train, cv=cv)
fin_result=np.mean(results)
`````

`````python
for i, _ in enumerate(results):
    print("{}th cross validation score: {}".format(i, _))
print("\ncross validation final score: {}".format(fin_result))
`````

`````python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(n_splits=10,shuffle=True,random_state=42)
results=cross_val_score(model, X_train, y_train, cv=cv)
fin_result=np.mean(results)
for i, _ in enumerate(results):
    print("{}th cross validation score: {}".format(i, _))
`````
