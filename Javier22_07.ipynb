```python
import pandas as pd
import numpy as np
import seaborn as sb
import missingno as msno
```

```python
ds = sb.load_dataset("titanic")
```

```python
# True = NULL
ds.isnull()
```

```python
# False = NULL
ds.notnull()
```

```python
#suma los NULL
ds.isnull().sum()
```

```python
#Por defecto es 0
ds.isnull().sum(axis = 0)
```

```python
msno.matrix(df=ds)
```
